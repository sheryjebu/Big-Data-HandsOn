```python
#Multivariate Analysis:
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
```

```python
sns.jointplot(x='Avg Session Length',y='Yearly Amount Spent',kind="scatter",data=cust_data)
```

```python
data1 = cust_data['Avg Session Length']
data2 = cust_data['Yearly Amount Spent']

corr, _ = pearsonr(data1, data2)
print('Persons correlation: %.2f' % corr)
```

```python
#The Persons  correlation is 0.36 . This shows that there is average correlation between Avg Session Length and Yearly Amount Spend
sns.jointplot(x='Time on Website',y='Yearly Amount Spent',kind="scatter",data=cust_data)
```

```python
data1 = cust_data['Time on Website']
data2 = cust_data['Yearly Amount Spent']

corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)
```

```python
#The Pearsons correlation is -0.003. This shows that there is no correlation between Time on Website and Yearly Amount Spend
sns.jointplot(x='Time on App',y='Yearly Amount Spent',kind="scatter",data=cust_data)
```

```python
data1 = cust_data['Time on App']
data2 = cust_data['Yearly Amount Spent']

corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.2f' % corr)
```

```python
#The Pearsons correlation is 0.5. This shows a strong and positive correlation between Time on App and Yearly Amount Spend
sns.jointplot(x='Length of Membership',y='Yearly Amount Spent',kind="scatter",data=cust_data)
```

```python
data1 = cust_data['Length of Membership']
data2 = cust_data['Yearly Amount Spent']

corr, _ = pearsonr(data1, data2)
print('Pearsons correlation: %.2f' % corr)
```

```python
#The Pearsons correlation is 0.81. This shows a strong and positive correlation between Length of Membership and Yearly Amount Spend
#Summary of correlation between variables by heatmap
sns.heatmap(cust_data.corr(),annot=True,cmap='Blues')
```

```python
#Interaction between Length of Membership and Yearly Amount Spent
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=cust_data)
```
